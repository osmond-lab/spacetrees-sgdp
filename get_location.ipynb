```python
import numpy as np
```

```python
input = ['data/SGDP.poplabels','data/SGDP_metadata.279public.21signedLetter.samples.txt']
```

```python
#pop_data = np.loadtxt(input[0],dtype=str)
#pop_data
#loc_data = np.loadtxt(input[1],dtype=str)
#loc_data

lonlats = []
with open(input[1],"rb") as f:
    next(f)
    for line in f:
        l = line.strip().split(b'\t')
        if l != [b'']:
            lonlat = []
            lonlat.append(l[4])
            lonlat.append(l[11])
            lonlat.append(l[12])
        lonlats.append(lonlat)
        
print(len(lonlats))
```

```python
type(lonlats)
```

```python
accessions = [i[0] for i in lonlats]
#print(accessions)
```

```python
IDs = []
with open(input[0],"rb") as f:
    next(f)
    for line in f:
        IDs.append(line.strip().split(b' ')[0])
        
#print(IDs)
```

```python
both = []
only_label = []
print('Length of poplabels: ',len(IDs))
print('Length of metadata: ',len(accessions))
for ID in IDs:
    try:
        index = accessions.index(ID)
        both.append(accessions[index])
    except ValueError:
        only_label.append(ID)
print("Elements only in poplabels: ", only_label)
print(len(IDs))

accessions_set = set(accessions)
both_set = set(both)
only_loc = accessions_set - both_set
only_loc_list = list(only_loc)

print("Elements only in metadata: ", only_loc_list)
print(len(accessions))
```

```python
order = []
for ID in IDs:
    ix = accessions.index(ID)
    order.append(ix)
locations = np.array([list(map(float,[i[2],i[1]])) for i in lonlats])[order]
print(locations)
print(type(locations))
```
